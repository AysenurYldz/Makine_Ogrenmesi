```python
df = df[(df['Quantity'] > 0)]
df = df[(df['Price'] > 0)]
```

```python
def rfm_table(dfm):
    rfm = dfm.groupby('Customer ID').agg({'InvoiceDate': lambda InvoiceDate: (today - InvoiceDate.max()).days,
                                         'Invoice': lambda Invoice: Invoice.nunique(),
                                         'TotalPrice': lambda TotalPrice: TotalPrice.sum()})
    
    rfm.columns = ['recency', 'frequency', 'monetary']

    return(rfm)
    
```

```python
rfm_tab = rfm_table(df)
```

```python
rfm_tab
```

```python
rfm_tab = rfm_tab[(rfm_tab['monetary']>0)]
```

```python
def rfm_score(dfm):
    seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'}
    
    dfm["recency_score"] = pd.qcut(dfm['recency'], 5, labels=[5, 4, 3, 2, 1])
    dfm["frequency_score"] = pd.qcut(dfm['frequency'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
    dfm["monetary_score"] = pd.qcut(dfm['monetary'], 5, labels=[1, 2, 3, 4, 5])
    dfm["RFM_SCORE"] = (dfm['recency_score'].astype(str) +
                        dfm['frequency_score'].astype(str))
    
    dfm['SEGMENT'] = dfm['RFM_SCORE'].replace(seg_map, regex=True)

    return(dfm)
```
